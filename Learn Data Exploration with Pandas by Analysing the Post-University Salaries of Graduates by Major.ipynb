### In this section we will be inserting a new column using insert()

## using the grouby() function

```python
# we want to group the data by groups columns
new_clean.groupby('Group').count()
```

##### pd.options.display.float_format = '{:,.2f}'.format  used to fromat the output 

```python
#now we can get the average salary for each column same as the standard deviation
pd.options.display.float_format = '{:,.2f}'.format 
new_clean.groupby('Group').mean()
```
